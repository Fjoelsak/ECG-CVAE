```python
import requests

url = "https://github.com/Fjoelsak/ECG-CVAE/releases/download/1.0.0/ST.npy"
out_path = "/content/ST.npy"

print("Downloading large NPY file (1.5GB)...")
r = requests.get(url, stream=True)
with open(out_path, "wb") as f:
    for chunk in r.iter_content(chunk_size=1024*1024):
        f.write(chunk)

print("Download complete:", out_path)
```
